```python
import re
import numpy as np
import pandas as pd
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")
warnings.simplefilter('ignore')
```

```python
def delete_useless(contents):

    case1 = re.search('.+?(?=[\.\s\(][a-zA-Z0-9]+?@)', str(contents))
    case2 = re.search('.+?(?=▶)', str(contents))
    case3 = re.search('.+?(?=\w{3}\s?기자)', str(contents))

    if case1 != None :
        contents = case1.group()
    elif case2 != None :
        contents = case2.group()
    elif case3 != None :
        contents = case3.group()
        
    case4 = re.search('기자 \= (.+)', str(contents))
    case5 = re.search('\](.+)', str(contents))
    
    if case4 != None :
        contents = case4.group(1)
    elif case5 != None :
        contents = case5.group(1)
        
    return contents
```

```python
def data_arrange_v1(theme):
    
    df = pd.read_csv("Data/theme_news_data/" + theme + ".txt", sep='\t', header=None, error_bad_lines=False, warn_bad_lines=False, engine='python')
    df = pd.DataFrame(df[3]).dropna().drop_duplicates().reset_index(drop=True)
    df.columns = ['contents']
    
    lst = []
    
    for i in range(len(df)):
        lst.append(delete_useless(df.contents[i]))
    
    for i in lst:
        if len(i) <= 100:
            lst.remove(i)
            
    df = pd.DataFrame(lst)
    df['theme'] = theme
    cols = df.columns.tolist()
    cols = cols[-1:] + cols[:-1]
    df = df[cols]
    df.columns = ['theme','contents']
    df = df[0:1000]
    
    return df
```
