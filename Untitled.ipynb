```python
import pandas as pd
```

```python
df1 = pd.read_csv("Data/filtered_data/final_data.csv",index_col=0)
df2 = pd.read_csv("Data/data_for_use/theme_contents_final.csv",index_col=0)
```

```python
df1
```

```python
lst = []
for i in range(len(df2)):
    lst.append(df2['theme'][i])

lst = list(set(lst))
lst = sorted(lst)
lst
```

```python
for i in range(len(df1)):
    if df1['Theme'][i] not in lst:
        df1 = df1.drop(index = i)

df1 = df1.reset_index(drop=True)
```

```python
df1.to_csv("Data/filtered_data/final_data_v1.csv")
```
