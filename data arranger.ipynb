```python
import re
import pandas as pd
import warnings
from tqdm import tqdm
import numpy as np
warnings.filterwarnings("ignore")
warnings.simplefilter('ignore')
```

```python
def delete_useless(contents):

    case1 = re.search('.+?(?=[\.\s\(][a-zA-Z0-9]+?@)', str(contents))
    case2 = re.search('.+?(?=▶)', str(contents))
    case3 = re.search('.+?(?=\w{3}\s?기자)', str(contents))

    if case1 != None :
        contents = case1.group()
    elif case2 != None :
        contents = case2.group()
    elif case3 != None :
        contents = case3.group()
        
    case4 = re.search('기자 \= (.+)', str(contents))
    case5 = re.search('\](.+)', str(contents))
    
    if case4 != None :
        contents = case4.group(1)
    elif case5 != None :
        contents = case5.group(1)
        
    return contents
```

```python
def data_arrange(theme):
    
    df = pd.read_csv("Data/theme_news_data/" + theme + ".txt", sep='\t', header=None, error_bad_lines=False, warn_bad_lines=False, engine='python')
    df = df[3]
    df = pd.DataFrame(df)
    df.columns = ['contents']
    df = df.dropna()
    df = df.drop_duplicates().reset_index()
    
    
    df = df[0:250]
    
    lst = []
    for i in range(len(df)):
        lst.append(delete_useless(df.contents[i]))
    for i in lst:
        if len(i) <= 50:
            lst.remove(i)
    df = pd.DataFrame(lst)
    df['theme'] = theme
    cols = df.columns.tolist()
    cols = cols[-1:] + cols[:-1]
    df = df[cols]
    df.columns = ['theme','contents']
    df = df[0:100]
    
    return df
```

```python
lst = pd.read_csv('Data/filtered_data/final_filtered_data_1.csv', index_col=0)
```

```python
for i in tqdm(range(len(lst.Theme.values))):
    try:
        if i == 0:
            df = data_arrange(lst.Theme.values[i])
        else:
            new_df = data_arrange(lst.Theme.values[i])
            df = pd.concat([df, new_df])
    except:
        print("Error occured: %s" % lst.Theme.values[i])
```

```python
df.to_csv('theme_contents_raw.csv')
```
