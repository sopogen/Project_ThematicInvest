```python
import re
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
warnings.simplefilter('ignore')
```

```python
def delete_useless(contents):

    case1 = re.search('.+?(?=[\.\s\(][a-zA-Z0-9]+?@)', str(contents))
    case2 = re.search('.+?(?=▶)', str(contents))
    case3 = re.search('.+?(?=\w{3}\s?기자)', str(contents))

    if case1 != None :
        contents = case1.group()
    elif case2 != None :
        contents = case2.group()
    elif case3 != None :
        contents = case3.group()
        
    case4 = re.search('기자 \= (.+)', str(contents))
    case5 = re.search('\](.+)', str(contents))
    
    if case4 != None :
        contents = case4.group(1)
    elif case5 != None :
        contents = case5.group(1)
        
    return contents
```

```python
df1 = pd.read_csv("Data/theme_news_data/2차전지.txt", sep='\t', header=None, error_bad_lines=False, engine='python')
df1.columns = ['data','newspaper','title','contents','link']
df1 = df1['contents']
df1 = df1.drop_duplicates().reset_index()
df1 = df1[0:200]
lst = []
for i in range(len(df1)):
    lst.append(delete_useless(df1.contents[i]))
for i in lst:
    if len(i) <= 50:
        lst.remove(i)
df1 = pd.DataFrame(lst)
df1['theme'] = '2차전지'
cols = df1.columns.tolist()
cols = cols[-1:] + cols[:-1]
df1 = df1[cols]
df1.columns = ['theme','contents']
df1 = df1[0:100]
df1
```
